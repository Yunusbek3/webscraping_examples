```python
# now collecting all the code into one place where it will be reshaped a little and stored
import requests as rd
from bs4 import BeautifulSoup
import pandas as pd

# This dictionary maps OLX category slugs (used in URLs) to readable English category names.
category = {
    'nedvizhimost': 'Real_Estate',
    'detskiy-mir': 'Kids_World',
    'transport': 'Transport',
    'zhivotnye': 'Animals',
    'dom-i-sad': 'Home_Garden',
    'elektronika': 'Electronical_applience',
    'uslugi': 'Services'
}

# A small utility function to turn a relative URL from OLX into a full, clickable link.
def convert_to_full_url(relative_url):
    return "https://www.olx.uz" + relative_url

# This list will gather all information from all categories to later store in a CSV file.
all_data = []

# Loop through each category slug (Uzbek version) in the dictionary.
for i in category:
    for j in range(1,26):
        # Build the search URL for the current category.
        url = f'https://www.olx.uz/oz/{i}/?currency=UZS&page={j}'
        Response = rd.get(url)
        soup = BeautifulSoup(Response.text, 'html.parser')

        # Find all product listings on the current category page.
        products = soup.find_all('div', class_="css-qfzx1y")

        # Loop through each product card on the webpage.
        for product in products:
            #getting the names of this product 
            defn = product.find('h4' , class_="css-1g61gc2").get_text()
            
            # Extract the price information shown in the ad.
            price_folder = product.find('p', class_='css-uj7mm0')
            price_converted = price_folder.get_text() if price_folder else "N/A"

            # Extract the product detail page URL to let users visit the original ad.
            link_tag = product.find('a', class_="css-1tqlkj0")
            relative_url = link_tag['href'] if link_tag else ""
            full_url = convert_to_full_url(relative_url)

            # Get the location and date when the ad was posted, shown at the bottom of the card.
            loc_date_tag = product.find('p', class_="css-vbz67q")
            loc_date = loc_date_tag.get_text() if loc_date_tag else "N/A"

            # Determine the condition or type of listing (e.g., new, good, or unspecified).
            div_tag = product.find('div', class_="css-krg4hw")
            if div_tag:
                span_tag = div_tag.find('span', class_="css-iudov9")
                condition = span_tag['title'] if span_tag and span_tag.has_attr('title') else "not specified"
            else:
                condition = "not specified"

            # Combine all collected data into a single row and label it with its English category name.
            all_data.append({
                'Price': price_converted,
                'URL': full_url,
                'Location_Date': loc_date,
                'ConditionCategory': condition,
                'MainCategory': category[i]  # Translates slug into readable category name
            })

# After collecting data from all categories, convert it into a DataFrame (table).
df = pd.DataFrame(all_data)

# Save the final table into a CSV file that can be used for analysis, visualization, etc.
df.to_csv("olx_scraped_data.csv", index=False, encoding='utf-8-sig')
print("CSV file saved as 'olx_scraped_data.csv'")
```
